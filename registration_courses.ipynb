```python
import pandas as pd
```

```python
df1=pd.read_csv('studentRegistration.csv')
df2=pd.read_csv('courses.csv')
```

```python
merged_df = pd.merge(df1, df2, on=["code_presentation", "code_module"], how="left")
merged_df.shape
```

```python
merged_df
```

```python
df=merged_df
absolute_values = df['date_registration'].abs()
```

```python
mode_value = absolute_values.mode()[0]
```

```python
 #Fill null values in the original column with the mode
df['date_registration'] = df['date_registration'].fillna(-mode_value)
```

```python
df.isnull().sum()
```

```python
import numpy as np
df['date_registration'] = df['date_registration'].abs()

# Calculate overallstay
df['overallstay'] = np.where(
    df['date_unregistration'].isna(),df['date_registration'] + df['module_presentation_length'],
    df['date_registration'] + df['date_unregistration']
)

df
```

```python
overallstay_column = df['overallstay']
```

```python

# Step 3: Add the extracted column back to merged_df (if needed)
merged_df['overallstay'] = overallstay_column
```

```python
merged_df['date_registration'] = merged_df['date_registration'] * -1
merged_df
```

```python
regiscourses=merged_df
```

```python
regiscourses=regiscourses.drop(['date_registration','date_unregistration','module_presentation_length'],axis=1)
```

```python
regiscourses = regiscourses.drop(['code_module', 'code_presentation'], axis=1)
```

```python
regiscourses.to_csv("regiscourses.csv", index=False)
```
